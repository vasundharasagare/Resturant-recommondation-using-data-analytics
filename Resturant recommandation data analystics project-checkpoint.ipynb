```python
def collapse(L):
    L1=[]
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
```

```python
rest['Cuisines']=rest['Cuisines'].apply(collapse)
rest['Locality']=rest['Locality'].apply(collapse)
rest['Locality Verbose']=rest['Locality Verbose'].apply(collapse)
rest['Address']=rest['Address'].apply(collapse)
```

```python
rest['City']=rest['City'].apply(lambda x:(x.replace(" ","")).split(" "))
```

```python
rest['tags']=rest['Cuisines']+rest['Locality']+rest['Locality Verbose']+rest['Address']
```

```python
new=rest[["Restaurant ID","Restaurant Name","Country Code","tags",'Has Table booking']]
```

```python
new.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vector = cv.fit_transform(new['tags']).toarray()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
```

```python
new[new['Restaurant Name'] == 'Le Petit Souffle'].index[0]
```

```python
def myrecommend(resto):
    restros=[]
    index = new[new['Restaurant Name']==resto].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key=lambda x: x[1])
    for i in distances[1:6]:
        restros.append((new.iloc[i[0]]['Restaurant Name']))
    return restros
```

```python
myrecommend('Le Petit Souffle')
```
